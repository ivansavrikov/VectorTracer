```python
from PIL import Image
from IPython.display import display, SVG
from core.BuilderSVG import raster_to_svg
import os

INPUT_DIR = './ResourcesForTesting/InputImages/'
OUTPUT_DIR = './ResourcesForTesting/Vectors/'

def temp_ui(image_path):
    image: Image = Image.open(image_path)
    display(image)
    svg_code = raster_to_svg(image)
    file_title = os.path.splitext(os.path.split(image_path)[1])[0]
    with open(f"{OUTPUT_DIR}{file_title}.svg", "w", encoding="utf-8") as svg:
        svg.write(svg_code)
    display(SVG(svg_code))

def main():
    images_paths: list = [f'{INPUT_DIR}center rect.jpg',
                          f'{INPUT_DIR}play.jpg',
                          f'{INPUT_DIR}romboplus.jpg',
                          f'{INPUT_DIR}3angle.jpg',
                          f'{INPUT_DIR}3angleup.jpg',
                          f'{INPUT_DIR}strange rect.jpg',
                          f'{INPUT_DIR}circle.jpg',
                          f'{INPUT_DIR}plus.jpg',
                          f'{INPUT_DIR}caban.bmp',
                          f'{INPUT_DIR}7angle.jpg',
                          f'{INPUT_DIR}heart.png']

    for p in images_paths:
        temp_ui(p)

if __name__ == '__main__':
    main()
```
